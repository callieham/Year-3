```python
import numpy as np
def generator(n):
    #u_i = np.random.uniform(-1, 1, n)
    u_i = np.linspace(-1, 1, n)
    noise = np.random.normal(0, 0.2, n)
    return u_i, u_i**2 + noise
```

```python
u, v = generator(200)
```

```python
import  matplotlib.pyplot  as plt
```

```python
plt.scatter(u, v, s = 0.5)
```

```python
S = np.transpose(np.array([u**2, u, np.ones(200)]))
```

```python
x_ls = np.dot(np.linalg.inv(np.dot(np.transpose(S), S)), np.dot(np.transpose(S), v))
```

```python
v_new = x_ls[0]*u**2 + x_ls[1]*u + x_ls[2]*np.ones(200)
plt.plot(u, v_new)
plt.scatter(u, v, s = 0.5)
```

```python
sum((v - v_new)**2)
```

```python
lam = 100
L = np.diag(np.ones(200)) - np.diag(np.ones(199), 1)
clean = np.dot(np.linalg.inv(np.identity(200) + lam*np.dot(np.transpose(L), L)), v)
```

```python
plt.plot(u, v_new)
plt.scatter(u, v, s = 0.5)
plt.plot(u, clean)
```
